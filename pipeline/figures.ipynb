# Coverage Profile

```python
import matplotlib.pyplot as plt

cov={}
count={}
cov=[0]*51
count=[0]*51
i =0 
summ =0 
for line in open("SRR606249.qc.coverage"):
            if i >50:
                    break
            x = line.rstrip().split(' ')
            cov[i]=int(x[0])
            count[i]=int(x[1])  
            i +=1


Y =count
X =cov 

# Compute the CDF
CY = np.cumsum(Y)
fig1 = plt.figure()
plt.plot(X, CY,color='#4863A0', linestyle='solid', linewidth=4, label ='Mapping Coverage')
plt.title('Coverage profile for the reference genome')
#plt.axis(xmax=50.0)
#plt.axis(ymax=1.0)
plt.xlabel('Coverage')
plt.ylabel('Fraction of reference at or above coverage')
fig1.savefig('CoverageProfile.png')
plt.show()
```

# Unalignments Histogram

```python
import matplotlib.pyplot as plt

fig2 = plt.figure()
ax = fig2.add_subplot(111)
## the data
N =3
 
iqc_aligned = 7067517/1000000.0  
iqc_mapped = 2332735/1000000.0  

sqc_aligned = 6786747/1000000.0  
sqc_mapped =  1966038/1000000.0  

mqc_aligned = 6392926/1000000.0  
mqc_mapped = 1585363/1000000.0  


incontigs = [iqc_aligned, sqc_aligned, mqc_aligned]
inreference = [iqc_mapped, sqc_mapped, mqc_mapped]

width = 0.35   
ind = np.arange(N)                # the x locations for the groups

# axes and labels
ax.set_xlim(-width,len(ind)+width)
ax.set_ylim(0,10)
ax.set_ylabel('No. of reads in millions')

## necessary variables
                   # the width of the bars

## the bars
rects1 = ax.bar(ind, incontigs, width,color='#4863A0')
rects2 = ax.bar(ind+width, inreference, width,color='#CCCCCC')

ax.set_title('Reads Mapped to Unaligned Contigs')
xTickMarks = ['IDBA','SPAdes','Megahit']
ax.set_xticks(ind+width)
xtickNames = ax.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=45, fontsize=9)

## add a legend
ax.legend( (rects1[0], rects2[0]), ('Reads Mapped', 'Reads that also map to reference') )

fig2.savefig('UnalignmentHistogram.png')
plt.show()
```
